###  2. Установить характер разрыва функции в точке x0:


1. f(x) = (x^2 - 16)/(x + 4), x0 = -4

Чтобы определить тип разрыва в точке x0 = -4 для функции f(x) = (x^2 - 16)/(x + 4), нам нужно исследовать левый и правый пределы функции в точке x0:

Левый предел:

lim x → -4- (x ^ 2 - 16)/(x + 4) = lim x → -4- [(x + 4)(x - 4)/(x + 4)] = lim x → -4 - (х - 4) = -8

Правый предел:

lim x → -4+ (x ^ 2 - 16)/(x + 4) = lim x → -4+ [(x + 4)(x - 4)/(x + 4)] = lim x → -4 + (х - 4) = -8

Поскольку и левый, и правый пределы существуют и равны, функция имеет конечный предел при x0 = -4, а именно -8.

Однако функция не определена при x0 = -4, так как знаменатель функции становится равным нулю, а значит, функция имеет вертикальную асимптоту при x0 = -4.

Следовательно, тип разрыва при x0 = -4 для функции f(x) = (x^2 - 16)/(x + 4) является вертикальной асимптотой или бесконечным разрывом.

2. f(x) = sin(x)/x, x0 = 0

Чтобы определить тип разрыва в точке x0 = 0 для функции f(x) = sin(x)/x, нам необходимо исследовать левый и правый пределы функции в точке x0:

Левый предел:

lim x → 0- sin(x)/x = lim x → 0- sin(x)/(-|x|) = -1

Правый предел:

lim x → 0+ sin(x)/x = lim x → 0+ sin(x)/|x| = 1

Поскольку левый и правый пределы не совпадают, предел функции не существует при x0 = 0. Следовательно, функция имеет разрыв при x0 = 0.

Этот тип разрыва называется устранимым разрывом, потому что функция имеет дыру в точке x0 = 0, которую можно заполнить, переопределив функцию в этой точке. Мы можем переопределить f(x) при x0 = 0, чтобы она была равна пределу функции, когда x приближается к 0, то есть:

lim x → 0 sin(x)/x = 1

Следовательно, мы можем переопределить f(x) при x0 = 0 так, чтобы она была равна 1, и функция станет непрерывной при x0 = 0.

Итак, характер разрыва при x0 = 0 для функции f(x) = sin(x)/x — разрыв устранимый.
