### 1. Решить уравнения:


1. y' − y = 2x − 3

Найдем множитель $\mu(x)$, умножив обе части уравнения на него:
$$\mu(x)y' - \mu(x)y = 2x\mu(x) - 3\mu(x)$$

Чтобы найти подходящий множитель $\mu(x)$, выберем его таким образом, чтобы левая часть полученного уравнения была полной производной некоторой функции, т.е.:
$$\mu(x)y' - \mu(x)y = (\mu(x)y)'$$

Для этого нужно найти функцию $\mu(x)$, удовлетворяющую уравнению:

$$\frac{d\mu(x)}{dx} = -\mu(x)$$

Которое имеет решение:

$$\mu(x) = e^{-x}$$

Теперь умножим исходное уравнение на множитель $\mu(x)$:
$$e^{-x}y' - e^{-x}y = 2xe^{-x} - 3e^{-x}$$

Заметим, что левая часть является производной произведения функций $e^{-x}$ и $y(x)$, поэтому:
$$(e^{-x}y(x))' = 2xe^{-x} - 3e^{-x}$$

Проинтегрируем обе части полученного уравнения:
$$e^{-x}y(x) = -2xe^{-x} - 3e^{-x} + C$$

где $C$ - произвольная постоянная интегрирования.

Разрешим уравнение относительно $y(x)$:
$$y(x) = -2x - 3 + Ce^{x}$$

где $C$ - произвольная постоянная.

2. x^2y' + xy + 1 = 0

Для решения данного уравнения воспользуемся методом Бернулли:

Перепишем уравнение в виде:
$$y' + \frac{1}{x}y = -\frac{1}{x^2}y^{-1}$$

Сделаем замену переменной $z = y^{-1}$. Тогда:
$$z' = -\frac{y'}{y^2}$$

и уравнение примет вид:

$$-xz' - z = -\frac{1}{x^2}$$

Умножим обе части уравнения на множитель интегрирующего множителя $\mu(x) = e^{-\ln{x}} = \frac{1}{x}$:
$$-\frac{z}{x} + z'\cdot 1 = -\frac{1}{x^3}$$

Левая часть является производной произведения функций $z(x)/x$, поэтому:
$$\left(\frac{z}{x}\right)' = -\frac{1}{x^3}$$

Проинтегрируем обе части полученного уравнения:
$$\frac{z}{x} = \frac{1}{2x^2} + C$$

где $C$ - произвольная постоянная интегрирования.

Возвращаясь к исходной переменной $y(x)$:
$$y(x) = \frac{1}{Cx} + \frac{1}{2x^2}$$

где $C$ - произвольная постоянная.

### 2.Исследовать ряд на сходимость:


1. $\sum\limits_{n=1}^\infty \frac{n+2}{n^2+n+1}$

Для исследования данного ряда на сходимость воспользуемся признаком сравнения:

$$\frac{n+2}{n^2+n+1} \leq \frac{n+2}{n^2} = \frac{1}{n} + \frac{2}{n^2}$$

Ряд $\sum\limits_{n=1}^\infty \frac{1}{n}$ - расходится (гармонический ряд).

Ряд $\sum\limits_{n=1}^\infty \frac{1}{n^2}$ - сходится (ряд известный как ряд Бернулли).

Таким образом, по признаку сравнения, ряд $\sum\limits_{n=1}^\infty \frac{n+2}{n^2+n+1}$ также расходится.

Ответ: ряд расходится.

2. $\sum\limits_{n=1}^\infty \frac{n^n}{n!}$

Для исследования ряда на сходимость воспользуемся признаком Даламбера:

$$\frac{a_{n+1}}{a_n} = \frac{(n+1)^{(n+1)}}{(n+1)!} \cdot \frac{n!}{n^n} = \frac{(n+1)^n}{n^n} = \left(1 + \frac{1}{n}\right)^n$$

Заметим, что $\lim\limits_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e$, где $e$ - константа Эйлера. Таким образом, по признаку Даламбера, ряд $\sum\limits_{n=1}^\infty \frac{n^n}{n!}$ может как сходиться, так и расходиться.

Для дополнительной проверки сходимости ряда воспользуемся признаком Коши:

$$\sqrt[n]{a_n} = \sqrt[n]{\frac{n^n}{n!}} = \frac{n}{(n!)^{1/n}}$$

Известно, что $\lim\limits_{n \to \infty} \frac{n}{(n!)^{1/n}} = 0$. Таким образом, по признаку Коши, ряд $\sum\limits_{n=1}^\infty \frac{n^n}{n!}$ сходится.

Ответ: ряд сходится.

3. $\sum\limits_{n=1}^\infty \ = \ n \cdot \left(1 - \frac{1}{n}\right)^{n^2}$


Для исследования ряда на сходимость воспользуемся признаком Коши:

$$\sqrt[n]{a_n} = \sqrt[n]{n \cdot \left(1 - \frac{1}{n}\right)^{n^2}} = \sqrt[n]{n} \cdot \left(1 - \frac{1}{n}\right)^n$$

Заметим, что $\lim\limits_{n \to \infty} \sqrt[n]{n} = 1$ и $\lim\limits_{n \to \infty} \left(1 - \frac{1}{n}\right)^n = \frac{1}{e}$. Тогда:

$$\lim_{n \to \infty} \sqrt[n]{a_n} = \lim_{n \to \infty} \sqrt[n]{n} \cdot \left(1 - \frac{1}{n}\right)^n = 1 \cdot \frac{1}{e} = \frac{1}{e} < 1$$

Таким образом, по признаку Коши, ряд $\sum\limits_{n=1}^\infty n \cdot \left(1 - \frac{1}{n}\right)^{n^2}$ сходится.

Ответ: ряд сходится.

4.$\sum\limits_{n=1}^\infty \frac{(-1)^{n+1}}{2n - \ln n}$ 

Для исследования ряда на сходимость воспользуемся признаком Лейбница. Согласно этому признаку, если последовательность $b_n = \frac{1}{2n - \ln n}$ убывает монотонно к нулю, то ряд $\sum\limits_{n=1}^\infty (-1)^{n+1} b_n$ сходится.

Для доказательства убывания монотонности последовательности $b_n$ рассмотрим:

$$\frac{b_{n+1}}{b_n} = \frac{2n - \ln n}{2n+2-\ln(n+1)} = \frac{2n - \ln n}{2n + 2 - (\ln n + \ln(1+\frac{1}{n}))}$$

$$= \frac{2n - \ln n}{2n + 2 - \ln n - \ln(1+\frac{1}{n})} = \frac{2n - \ln n}{2n + 2 - \ln n - \frac{1}{n} + O(\frac{1}{n^2})}$$

При $n \geq 1$ последний знаменатель всегда больше числителя, так как $\ln n > 1/n$ при $n \geq 2$, а при $n=1$ неравенство также выполнено. Следовательно, $\frac{b_{n+1}}{b_n} < 1$ для всех $n \geq 1$, что означает убывание монотонности последовательности $b_n$.

Таким образом, по признаку Лейбница, ряд $\sum\limits_{n=1}^\infty \frac{(-1)^{n+1}}{2n - \ln n}$ сходится.

Ответ: ряд сходится.

5. $\sum\limits_{n=1}^\infty \frac{(x-2)^{n+1}}{3^n(n+2)}$ 

Для исследования ряда на сходимость воспользуемся признаком Даламбера:

$$\lim_{n \to \infty} \frac{a_{n+1}}{a_n} = \lim_{n \to \infty} \frac{(x-2)^{n+2}}{3^{n+1}(n+3)} \cdot \frac{3^n(n+2)}{(x-2)^{n+1}} = \lim_{n \to \infty} \frac{(x-2)}{3} \cdot \frac{n+2}{n+3} = \frac{(x-2)}{3}$$

Таким образом, ряд сходится при $\frac{(x-2)}{3} < 1$, то есть при $x < 5$. Абсолютная сходимость ряда будет обеспечена при:

$$\sum_{n=1}^{\infty} \left|\frac{(x-2)^{n+1}}{3^n(n+2)}\right| = \sum_{n=1}^{\infty} \frac{|x-2|^{n+1}}{3^n(n+2)} < \infty$$

Для доказательства абсолютной сходимости ряда рассмотрим признак Коши:

$$\lim_{n\to\infty}\sqrt[n]{\frac{|x-2|^{n+1}}{3^n(n+2)}} = \lim_{n\to\infty}\frac{|x-2|}{3}\sqrt[n]{\frac{1}{n+2}} = \frac{|x-2|}{3} < 1$$

Следовательно, при $|x-2| < 3$, ряд сходится абсолютно.

Таким образом, ряд сходится при $x \in (2,5)$ и абсолютно сходится при $x \in (2-3,2+3) = (-1,5)$.

Ответ: ряд сходится при $x \in (2,5)$ и абсолютно сходится при $x \in (-1,5)$.
